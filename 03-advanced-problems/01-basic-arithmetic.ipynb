#### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
import cmath

print("Enter the co-ordinates of point-1 (p1x, p1y): ")
p1x, p1y = int(input("p1x = ")), int(input("p1y = "))

print("Enter the co-ordinates of point-2 (y1, y2): ")
p2x, p2y  = int(input("p2x = ")), int(input("p2y = ")),

d = cmath.sqrt( (p1x-p2x)**2 + (p1y-p2y)**2 )
print(f"Distance between two points = {round(d.real,2)}") 
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
first_term = int(input("Enter 1st term: "))
second_term = int(input("Enter 2nd term: "))
n = int(input("Enter the `nth` term to find: "))
a = first_term 
d = second_term -  first_term

an = a + (n-1) * d
print(f"{n}th term of serise = {an}")
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here

print("Enter numerator and denominator of 1st number (n1/d1): ")
n1, d1 = int(input("n1 = ")), int(input("d1 = "))

print("Enter numerator and denominator of 2st number (n2/d2): ")
n2, d2 = int(input("n2 = ")), int(input("d2 = "))

fraction_result = (d2 * n1 + d1 * n2 ) / (d1 * d2)
print(f"{n1}/{d1} + {n2}/{d2} = {round(fraction_result, 2)}")
````
