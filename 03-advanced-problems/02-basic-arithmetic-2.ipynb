#### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Write code here 
# ctc = int(input('Enter your annual CTC in lakhs: '))
ctc  = 12
deduction_rate = 0.18
if ctc < 5:
  tax_rate = 0.0

elif ctc < 10:
  tax_rate = 0.10

if ctc < 20:
  tax_rate = 0.20

else:
  tax_rate = 0.30
 
remaining_salary  = ctc * (1-deduction_rate)
annual_salary = remaining_salary * (1 - tax_rate)
monthly_salary = annual_salary * 100000 / 12
print(f"In hand monthly salary for {ctc} Lakh ctc = {round(monthly_salary, 2)}")
````

#### `Problem 5` - Exercise 12: Display Fibonacci series up to 10 terms.

*Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34*

````python
# Write code here
count, first, second, next = 0, -1, 1, 0

print("Fibonacci Serise:")
while count<10 :
  next = first + second
  print(f"{next}", end=" ")
  first = second
  second = next
  count += 1
````

#### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
# Write code here
num = int(input("Enter a number: "))
i = num
fact = 1

while i >= 1 :
  fact *= i  
  i -= 1
print(f"{num}! = {fact}")
````

#### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Write code here
n = int(input("Enter a number: "))
sum, i = 0, 1

while i <= n :
  if i % 5 != 0 :
    sum += i
 
  if sum > 300 :
    break  
  i += 1

print(f"{sum}")  
````

````python
# Write code here
sum = avg = counter = 0
while True :
  n = int(input("Enter a number: "))
  if n == 0:
    break
  counter += 1
  sum += n
if counter == 0:   
  print(f"Sum = {sum}\nAverage = {avg}")
else :
  print(f"Sum = {sum}\nAverage = {sum / counter}")  
````

#### `Problem 11`: Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.



````python
# Write code here
even = "02468"
for e in range(1000, 3000) :
  counter = 0
  temp = e
  while temp :
    if str(temp % 10) in even :
      counter += 1
    temp //= 10  

  if counter >= 4 :
    print(e,end='-')  
````

```

LEFT 3

```

> `!` means robot stop there.

Example:

`Input`:
```

```

#### `Problem 13`:Write a program to print whether a given number is a prime number or not

````python
# Write code here
n = int(input("Enter a number: "))

if n < 2 :
  print(f"{n} is not a prime number.")
else :
  for e in range(2,n+1):
    if n % e == 0:
      break

if e == n :
    print(f"{n} is a prime number.")
else:  
    print(f"{n} is not a prime number.")    
````

#### `Problem 14`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# Write code here
start = int(input("Enter start value of range: "))   
end = int(input("Enter end value of range: ")) 

for e in range(start, end) :
  arm = no_of_digit = 0 
  temp = e

# count no of digits
  while temp :
    temp //= 10
    no_of_digit += 1

  temp = e  

# Sum of digits    246
  while temp :    
    digit = temp % 10
    arm += digit ** no_of_digit
    temp //= 10

  if arm == e : 
    print(f"{e}", end=' ')  
  
````
