### 12. Write a class 'SmartPhone' with Getter and Setter methods.
> **Logic/Explanation:** Getters retrieve private data; Setters update it. This allows you to add 'validation logic' (e.g., price can't be negative) before changing data.


```python
# write your code here 
class Smartphone:
  def __init__(self):
    self.__price = 0
    self.__brand = ''
    self.__storage = ''

  def set_details(self, brand: str, price: int, storage: str) -> None:
    if type(price) != int:
      print('Please enter correct integer value of price.')
      return
    else:  
      self.__price += price

    self.__brand += brand
    self.__storage += storage

  def get_details(self) -> None:
    if self.__brand == '' and self.__storage == '':
      return 
    else:
      print(f'Brand : {self.__brand}')  
      print(f'Price : {self.__price}')  
      print(f'Storage : {self.__storage}')  

s = Smartphone()    
s.set_details('Apple', 80000, '12 GB')
s.get_details()
```

### 14. Create a class 'Temperature' using the '@property' decorator.
> **Logic/Explanation:** The `@property` decorator allows you to call a method like it's a simple variable (e.g., `obj.temp` instead of `obj.get_temp()`).


```python
# write your code here 
class Temperature:

    def __init__(self, celsius):
        self._celsius = celsius


    @property
    def celsius(self):
        return self._celsius


    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            print("Temperature cannot be below absolute zero!")
        else:
            self._celsius = value


t1 = Temperature(25)

print(t1.celsius)

t1.celsius = 40
print(t1.celsius)
```

### 17. Create a Multilevel Inheritance (Vehicle -> Car -> ElectricCar).

```python
# write your code here 
class  Vehicle:
  def __init__(self, wheel: int, seat: int):
    self.weels = wheel
    self.seater = seat

  def vehicle_info(self):
    print(f'This is a {self.seater} seater vehile and {self.weels} weeler')

class Car(Vehicle):
  engine = '300 HP'
  
  def car_info(self):
    print(f'This car has {self.engine} engine')

class ElectricCar(Car):
  battery = 10
  
  def electric_car_info(self):
    print(f'This car has {ElectricCar.battery} battries')


ec = ElectricCar(4, 5)
ec.vehicle_info()
ec.car_info()
ec.electric_car_info()
```

### 29. Overload '<' (__lt__) to compare Rectangles.
> **Logic/Explanation:** Define `__lt__` (less than) to compare objects based on a specific logic (like area).


```python
# write your code here 
class Rectangle:
  def __init__(self, l: int, b:int):
    self.length = l
    self.breadth = b

  def area(self):
    self.area =  self.length * self.breadth
    return self.area
  
  def __lt__(self, other):
    if self.area < other.area :
      return True
    else:
      return False


r1 = Rectangle(10,2)   
r1.area()  

r2 = Rectangle(5,2)   
r2.area()  

r1 > r2 
```

### 30. Overload 'len()' for a custom object.

```python
# write your code here 
class Rectangle:
  def __init__(self, l: int, b:int):
    self.length = l
    self.breadth = b

  def area(self):
    self.area =  self.length * self.breadth
    return self.area
  
  def __lt__(self, other):
    if self.area < other.area :
      return True
    else:
      return False

  def __len__(self):
    return self.length

r1 = Rectangle(10,2)   
r1.area()  

r2 = Rectangle(5,2)   
r2.area()  

r1 > r2 

print(f'Length of r1 = {len(r1)}')
print(f'Length of r2 = {len(r2)}')
```

### 32. Create abstract 'Bank' class using 'abc' module.
> **Logic/Explanation:** Import `ABC` and `@abstractmethod` from the `abc` module to define rules that child classes **must** follow.


```python
# write your code here 
from abc import ABC, abstractmethod

class Bank(ABC):
  def __init__(self):
    self.balance = 0

  def connection(self):
    print('Connected to database.')  

  def bank_info(self):
    print('Welcome to Banking application.')  

  @abstractmethod
  def security(self):
    pass  

class MobileBanking(Bank):
   
   def security(self):
     print('Login registration security implemented.')

mb = MobileBanking()
mb.connection()
mb.security()
```
