#### 1. Basic Try-Except Block: Write a program that asks the user for a number and attempts to divide 100 by that number. Use a try-except block to catch a ZeroDivisionError and print a helpful message if the user enters 0.


```python
# write your code here
try:
  num = int(input('Enter a number'))
  result = 100 / num
  print(f'100 / {num} = {result}')
except ZeroDivisionError as e:
  print('You can n not divide any number by 0')
```
