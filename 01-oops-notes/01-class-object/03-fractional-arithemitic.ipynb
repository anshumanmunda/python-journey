## Operator Overloading
>Creating a custom class for arithmetic operation between instence objects

```python
class Fraction:
  def __init__(self, num:int, deno:int) -> None:  
    self.num = num
    self.deno = deno

  def show_result(self) -> None:
    print(f'{self.num}/{self.deno}')  

  def __add__(self, other):
    numerator = other.deno * self.num + self.deno * other.num 
    denominator = self.deno * other.deno
    return Fraction(numerator, denominator)

f1 = Fraction(2,3)
f2 = Fraction(4,3)

f3 = f1 + f2
f3.show_result()
```
