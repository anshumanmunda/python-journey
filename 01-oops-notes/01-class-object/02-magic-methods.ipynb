>Let's Perform operation in **user-definde** class

```python
class Line:
  def __init__(self, l):
    self.length = l

  def show_length(self): 
    print(self.length) 

l1 = Line(10)
l2 = Line(20)
l3 = l1 + l2
```

>**In order to perform arithmetic addition we need to define magic method in our class**

`__add__()`

```python
class Line:
  def __init__(self, l):
    self.length = l

  def show_length(self): 
    print(f'Length = {self.length}') 

  def __add__(self, other):
    new_length = self.length + other.length
    return Line( new_length )
  
  def __sub__(self, other):
    new_length = self.length - other.length
    return Line( new_length )
    

l1 = Line(10)
l2 = Line(20)

l3 = l1 + l2
l3.show_length()

l4 = l1 - l2
l4.show_length()
```
