## Name Conflicts

### **Conflict - 1** : Instance Object Variable Name Conflict

```python
class A:
  def __init__(self):
    self.a = 10

class B(A):
  def __init__(self):
    super().__init__()
    self.a = 20
    
```

```python
obj = B()
obj.a
```

### **Conflict - 2** : Class Object (Static) Variable Name Conflict

```python
#  CASE-1: If child class don't have the variable.

class A:
  x  = 10

class B(A):
  y = 20

print(A.x)
print(B.x)
```

```python
# If the child class defines a variable with the same name as the parent.
class A:
  x  = 10

class B(A):
  x = 20

print(A.x)
print(B.x)
```

### **Conflict - 3** : Instance Method Name Conflict

```python
# Method overriding 
class A:
  def fun(self):
    print('Class A function got executed.')

class B(A):
  def fun(self):
    print('Class B function got executed.')

obj = B()
obj.fun()
```

```python
# Method Hiding
class A:
  def area(self, l, b):
    return l * b
  
class B(A):
  def area(self, side:int):
    return side ** 2

obj = B()
result = obj.area(7)
print(f'Area  = {result}')
```

```python
class A:
  def f1(self):
    print('First Function')

  def f1(self):
    print('Second Function')

```

### **Conflict - 4** : Static Method

```python
class A:
  @staticmethod
  def f1():
    print('This is class A')

class B(A):
  @staticmethod
  def f1():
    print('This is class B')    

A.f1()   
B.f1()   
```

### **Conflict - 5** : Class Method

```python
class A:
  @classmethod
  def f1(cls):
    print('This is class A')

class B(A):
  @classmethod
  def f1(cls):
    print('This is class B')    

A.f1()   
B.f1()   
```
