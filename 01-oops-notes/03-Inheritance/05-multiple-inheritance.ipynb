### Type- 4: Multiple Inheritence

```python
class Product:
  def buy(self):
    print('Buying a Product.')

class Phone:
  def review(self):
    print('Reviewing a Phone.')

class Smartphone(Product, Phone):
  def __init__(self, brand: str, price: int):
    self.brand = brand
    self.price = price

  def get_info(self):
    print(f'Brand: {self.brand}')  
    print(f'Price: {self.price}')  
```

```python
s = Smartphone('Samsung', 70000)
s.get_info()
s.review()
s.buy()
```
