### Type 5: Hybrid Inheritance

```python
class University:
  def info(self):
    print('This is a University.')

class Faculty(University):
  def faculty_info(self):
    print('This is Faculty department.')

class Research(University):
  def research_info(self):
    print('This is reasearch department.')

class Project:
  def project_info(self):
    print('This is project details. ')

class Student(Faculty, Project):
  def student_info(self):
    print('Project Completed by student ')
```

```python
s = Student()
s.info()
s.faculty_info()
s.student_info()
```
