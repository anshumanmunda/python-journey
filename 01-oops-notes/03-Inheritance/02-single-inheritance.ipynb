## Types of Inheritance


### Type 1:  Single Inheritance

```python
class Phone():
  def __init__(self, price: int, brand: str, camera: str):
    self.brand = brand
    self.price = price
    self.camera = camera
  
  def buy(self):
    print('Buying Phone.')

class Smartphone(Phone):
  def __init__(self,  brand: str, price: int, camera: str, refresh_rate: str):
    super().__init__(price, brand, camera)
    self.refresh_rate = refresh_rate

  def get_info(self):
    print(f'Brand: {self.brand }') 
    print(f'Price: {self.price }') 
    print(f'Camera: {self.camera }') 
    print(f'Refresh Rate: {self.refresh_rate }') 
  
  def smart(self):
    print('Buying Phone.')
    print('You are in Phone class.')
  
```

```python
s1 = Smartphone('One Pluse', 60000, '60 MP', '120 HZ')
s1.buy()
s1.get_info()
```
