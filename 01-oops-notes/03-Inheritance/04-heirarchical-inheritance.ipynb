### Type 3 : Heirarchical Inheritance

```python
class Person:
  def __init__(self, name:str):
    self.name = name

  def display_person(self):
    print(f'{self.name} is a person.')  

class Employee(Person):
  def display_emp(self):
    print(f'{self.name} is a employee.')  

class Student( Person):
  def display_std(self):
    print(f'{self.name} is a student.')  
```

```python
s = Student('Anshuman')
s.display_person()
s.display_std()
```

```python
e = Employee('Rahul')
e.display_person() 
e.display_emp()
```
