## Inheritance

### Person and Student class Demonstration

```python
class Person:

  def __init__(self, name: str, age: int)-> None:
    self.name = name
    self.age = age
  
  def func(self):
    print(f'{self.name} is a person 🙍‍♂️')

class Student(Person):

  def set_details(self, roll: int, dept: str) -> None:
    self.roll_no = roll    
    self.dept = dept   
  

  def get_details(self):
    print(f'Name -> {self.name}')  
    print(f'Age -> {self.age}')  
    print(f'Roll no -> {self.roll_no}')   
    print(f'Department -> {self.dept}')   
   
```

```python
s = Student('Anshuman', 22)
s.set_details(1, 'MCA')
s.get_details()
print()

s2 = Student('Manish', 24)
s2.set_details(2, 'BCA')
s2.get_details()
```

## Constructor overriding 

```python
class Person:

  def __init__(self, name: str, age: int)-> None:
    self.name = name
    self.age = age
  
  def func(self):
    print(f'{self.name} is a person 🙍‍♂️')

class Student(Person):

  def __init__(self, name: str, age: int, roll: int, dept: str):
    # Person.__init__(self, name, age)
    super().__init__(name, age)
    self.roll_no = roll
    self.dept = dept
    
  
  def get_details(self):
    print(f'Name -> {self.name}')  
    print(f'Age -> {self.age}')  
    print(f'Roll no -> {self.roll_no}')   
    print(f'Department -> {self.dept}')   
   
```

```python
# s1 = Student(100, 'MCA')  # -> __init__(s1, 100, 'MCA)

s1 = Student('Anshuman', 22, 100, 'MCA')
s1.get_details()
```
