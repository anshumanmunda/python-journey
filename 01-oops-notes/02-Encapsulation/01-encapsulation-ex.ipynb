# Encapsulation

>Example Scenario: In an ATM application, if the balance attribute is public, an external user or a "malicious" junior programmer could set the balance to a string (e.g., "hehehe") instead of an integer. This causes the program to crash when a mathematical operation (like a withdrawal) is attempted because you cannot subtract an integer from a string

```python
class ATM:
  def __init__(self):
    self.pin = ''
    self.balance = 0

  def create_pin(self):
    if self.pin == '':
      user_pin = input('Please enter your pin: ')
      self.pin += user_pin
      print('Pin created successfully.') 
    else:
      print('Pin already created.')    

  def change_pin(self):
    if self.pin != '':
      old_pin = input('Enter your old pin: ')
      if old_pin == self.pin :
        self.pin = ''
      else:
        print('Your pin is incorrect.')
    else:
      print('Please create your pin first.')
      self.create_pin()      

  def deposit(self):
    if self.pin != '':
      amount = int(input('Enter your amount: '))
      self.balance += amount
      print(f'₹{amount} added successfully.') 
    else:
      print('Please create your pin first.')     


  def withdraw(self):
    if self.pin != '':
      if self.balance > 0 :
        amount = int(input('Enter the amount to withdraw: '))
        if amount > self.balance:
          print('Insufficent Balance.')

        else:  
          self.balance -= amount
          print(f'₹{amount} withdraw successfull.')
          print(f'Your avaliable balance = {self.check_balance()}')
      else:
        print(f'Your avaliable balance is to low, You can not withdraw money.')    

    else:
      print('Please create your pin first.')

  def check_balance(self):
    if self.pin != '':
      print(f'Your avaliable balance is ₹{self.balance}')

    else:
      print('Please create your pin first.')
   
```

> Let's make our balance attribute **Private** inside the class

```python
class ATM:
  def __init__(self):
    self.pin = ''
    self.__balance = 0

  def create_pin(self):
    if self.pin == '':
      user_pin = input('Please enter your pin: ')
      self.pin += user_pin
      print('Pin created successfully.') 
    else:
      print('Pin already created.')    

  def change_pin(self):
    if self.pin != '':
      old_pin = input('Enter your old pin: ')
      if old_pin == self.pin :
        self.pin = ''
      else:
        print('Your pin is incorrect.')
    else:
      print('Please create your pin first.')
      self.create_pin()      

  def deposit(self):
    if self.pin != '':
      amount = int(input('Enter your amount: '))
      self.__balance += amount
      print(f'₹{amount} added successfully.') 
    else:
      print('Please create your pin first.')     


  def withdraw(self):
    if self.pin != '':
      if self.__balance > 0 :
        amount = int(input('Enter the amount to withdraw: '))
        if amount > self.__balance:
          print('Insufficent Balance.')

        else:  
          self.__balance -= amount
          print(f'₹{amount} withdraw successfull.')
          print(f'Your avaliable balance = {self.check_balance()}')
      else:
        print(f'Your avaliable balance is to low, You can not withdraw money.')    

    else:
      print('Please create your pin first.')

  def check_balance(self):
    if self.pin != '':
      print(f'Your avaliable balance is ₹{self.__balance}')

    else:
      print('Please create your pin first.')
   
```

```python
# Creating an object
obj2 = ATM()

# Trying to access the private attribute balance
obj2.__balance
```

```python
# Trying to alter the private attribute
obj2.__balance = "hehe" 

# Even if we try to make changes. Inside our private attribute It will not crash our programme Because trying to do such changes creates New attribute
```

```python
obj2.create_pin()
```

```python
obj2.deposit()
```

```python
obj2.check_balance()
```

Private values are handled using `Name mangling` <br>
