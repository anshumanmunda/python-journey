## Abstraction

```python
from abc import ABC, abstractmethod

class BankApp(ABC):

  def database_connect(self):
    print('Database connected')

  @abstractmethod
  def security(self):
    pass

  @abstractmethod
  def display(self):
    pass


class MobileApp(BankApp):
  def mobile_login(self):
    print('Login to mobile app')

  def security(self):
    print('Mobile fingerprint security implemented')  

  def display(self):
    print('Displaying mobile interface')  
    
```

```python
obj = MobileApp()
obj.database_connect()
obj.security()
obj.display()
```
