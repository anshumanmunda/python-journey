## 1. Method overriding

```python
class Phone:
  def buy(self):
    print('Buying a phone')

class Smartphone(Phone):
    def buy(self):
      print('Buying a smart phone')

s = Smartphone()
s.buy()
```

## 2. Operator Overloading

```python
class Line:
  def __init__(self, length):
    self.length = length

l1 = Line(10)
l2 = Line(20)
l3 = l1 + l2    
```

```python
class Line:
  def __init__(self, length):
    self.length = length

  def __add__(self, other):
    sum = self.length + other.length
    return Line(sum)

l1 = Line(10)
l2 = Line(20)
l3 = l1 + l2  
print(l3.length)  
```
